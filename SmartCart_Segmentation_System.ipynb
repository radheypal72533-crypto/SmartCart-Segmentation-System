Handling Missing Values

```python
df['Income'] = df['Income'].fillna(df['Income'].median())
```

Feature Engeneering

```python
df['Age'] = 2026 - df['Year_Birth']
```

```python
df.head()
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'],dayfirst=True)
df['Customer_Join_Days'] = (df['Dt_Customer'].max() - df['Dt_Customer']).dt.days
```

```python
df.head()
```

```python
df.columns
```

```python
df['Total_spending'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
```

```python
df.head()
```

```python
df['Total_Children'] = df['Kidhome'] + df['Teenhome']
```

```python
df.head()
```

```python
df['Education'].value_counts()
```

```python
df['Education'] = df['Education'].replace({
  'Graduation' : 'Graduate',
  'Master' : 'Post Graduate',
  'Phd' : 'Post Graduate',
  '2n Cycle' : 'Ungraduate',
  'Basic' : 'Ungraduate'
})
```

```python
df['Marital_Status'].value_counts()
```

```python
df['Living_with'] = df['Marital_Status'].replace({
  'Married' : 'Living',
  'Together' : 'Living',
  'Single' : 'Alone',
  'Divorced' : 'Alone',
  'Widow' : 'Alone',
  'Absurd' : 'Alone',
  'YOLO' : 'Alone'
})
```

```python
df.head()
```

Encoding

```python
from sklearn.preprocessing import OneHotEncoder
cat_columns = df.select_dtypes(include='object').columns
one = OneHotEncoder(handle_unknown='ignore',sparse_output=False,drop='first')
new = one.fit_transform(df[cat_columns])
df2 = pd.DataFrame(new,columns=one.get_feature_names_out(),index=df.index)
df = pd.concat((df.drop(cat_columns,axis=1),df2),axis=1)
```

```python
df.head()
```

Standard Scaler

```python
X = df
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_scaled = sc.fit_transform(X)
```

Visualize The Data

```python
from sklearn.decomposition import PCA 
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)
```

```python
pca.explained_variance_ratio_
```

```python
fig = plt.figure(figsize=(8, 6))

ax = fig.add_subplot(111, projection="3d")

ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2])

ax.set_xlabel("PCA1")
ax.set_ylabel("PCA2")
ax.set_zlabel("PCA3")
ax.set_title("3d projection")
```

```python
from sklearn.cluster import KMeans
from kneed import KneeLocator
wcss = []
for k in range(1,11):
  kmeans = KMeans(n_clusters=k,random_state=42)
  kmeans.fit_predict(X_pca)
  wcss.append(kmeans.inertia_)
```

```python
knee = KneeLocator(range(1,11),wcss,curve='convex',direction='decreasing')
knee.elbow
```

```python

sns.lineplot(x=range(1,11),y=wcss,marker='o')
plt.xlabel("K")
plt.ylabel("WCSS")
```

```python
from sklearn.metrics import silhouette_score
scores = []
for k in range(2,11):
  kmeans = KMeans(n_clusters=k,random_state=42)
  labels = kmeans.fit_predict(X_pca)
  score = silhouette_score(X_pca,labels)
  scores.append(score)
```

```python
sns.lineplot(x=range(2,11),y=scores,marker='o')
```

Clustering

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
labels = kmeans.fit_predict(X_pca)
```

```python
fig = plt.figure(figsize=(8, 6))

ax = fig.add_subplot(111, projection="3d")

ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2],c=labels)

ax.set_xlabel("PCA1")
ax.set_ylabel("PCA2")
ax.set_zlabel("PCA3")
ax.set_title("3d projection")
```

```python
from sklearn.cluster import AgglomerativeClustering
ag = AgglomerativeClustering(n_clusters=3)
labels = ag.fit_predict(X_pca)
```

```python
fig = plt.figure(figsize=(8, 6))

ax = fig.add_subplot(111, projection="3d")

ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2],c=labels)

ax.set_xlabel("PCA1")
ax.set_ylabel("PCA2")
ax.set_zlabel("PCA3")
ax.set_title("3d projection")
```

```python
df['Cluster'] = labels
```
